```python
import numpy
import pandas
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
digits.data[0]
```

```python
dir(digits)
```

```python
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(digits.images[0])
```

```python
#data[0] shows pixels of image1 whereas image show image
for i in range(10):
    plt.matshow(digits.images[i])
    plt.show()
```

```python
X=digits.data
y=digits.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([digits.data[67]])
```

```python
plt.matshow(digits.images[67])
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_predicted=model.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_predicted))
cm=confusion_matrix(y_test,y_predicted)
import seaborn as sns
sns.heatmap(cm,annot=True)
plt.xlabel("predicted")
plt.ylabel("Truth")
plt.show()
#see that in truth of 1 and predicted of 3 there is value 1 which represents it predicted 1 as 3 1 times,value in that box contains how many times it contained that.
```

Let's build SVM model

```python
from sklearn.svm import SVC
clf=SVC(C=1000) #C is an regularization
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_predicted=clf.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_predicted))
cm=confusion_matrix(y_test,y_predicted)
import seaborn as sns
sns.heatmap(cm,annot=True)
plt.xlabel("predicted")
plt.ylabel("Truth")
plt.show()
#
```

SVM performs better than Logistic regression with a maximum accuracy of 99% while logistic performs 96%.So SVM is better when the target variable is multiclass and image classofication for these kind of simple classification tasks.
